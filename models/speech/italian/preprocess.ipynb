```python
import os
from pathlib import Path

import librosa
import numpy as np
import pandas as pd
import parselmouth
from tqdm import tqdm
```

```python
def stats(x, prefix):
    return {
        f"{prefix}_mean": np.mean(x),
        f"{prefix}_std": np.std(x),
        f"{prefix}_median": np.median(x),
        f"{prefix}_min": np.min(x),
        f"{prefix}_max": np.max(x),
    }
```

```python
def extract_parselmouth_features(wav_path):
    features = {}
    try:
        snd = parselmouth.Sound(wav_path)
        pitch = snd.to_pitch()
        pitch_values = pitch.selected_array["frequency"]
        pitch_values = pitch_values[pitch_values > 0]

        if len(pitch_values):
            features["pitch_mean"] = np.mean(pitch_values)
            features["pitch_std"] = np.std(pitch_values)
            features["pitch_min"] = np.min(pitch_values)
            features["pitch_max"] = np.max(pitch_values)

        else:
            features["pitch_mean"] = np.nan
            features["pitch_std"] = np.nan
            features["pitch_min"] = np.nan
            features["pitch_max"] = np.nan

        point_process = parselmouth.praat.call(
            snd,
            "To PointProcess (periodic, cc)",
            75,
            500
        )

        features["jitter_local"] = parselmouth.praat.call(
            point_process,
            "Get jitter (local)",
            0,
            0,
            0.0001,
            0.02,
            1.3
        )

        features["jitter_rap"] = parselmouth.praat.call(
            point_process,
            "Get jitter (rap)",
            0,
            0,
            0.0001,
            0.02,
            1.3
        )

        features["jitter_ppq5"] = parselmouth.praat.call(
            point_process,
            "Get jitter (ppq5)",
            0,
            0,
            0.0001,
            0.02,
            1.3
        )

        features["shimmer_local"] = parselmouth.praat.call(
            [snd, point_process],
            "Get shimmer (local)",
            0,
            0,
            0.0001,
            0.02,
            1.3,
            1.6
        )

        features["shimmer_apq3"] = parselmouth.praat.call(
            [snd, point_process],
            "Get shimmer (apq3)",
            0,
            0,
            0.0001,
            0.02,
            1.3,
            1.6
        )

        features["shimmer_apq5"] = parselmouth.praat.call(
            [snd, point_process],
            "Get shimmer (apq5)",
            0,
            0,
            0.0001,
            0.02,
            1.3,
            1.6
        )

        harmonicity = parselmouth.praat.call(
            snd,
            "To Harmonicity (cc)",
            0.01,
            75,
            0.1,
            1.0
        )

        features["hnr"] = parselmouth.praat.call(
            harmonicity,
            "Get mean",
            0,
            0
        )

    except Exception:

        pass

    return features
```

```python
def extract_librosa_features(y, sr):
    features = {}
    # -----------------------------
    # MFCC
    # -----------------------------
    mfcc = librosa.feature.mfcc(
        y=y,
        sr=sr,
        n_mfcc=13
    )

    for i in range(13):
        features.update(
            stats(
                mfcc[i],
                f"mfcc_{i+1}"
            )
        )

    # -----------------------------
    # Spectral Centroid
    # -----------------------------

    centroid = librosa.feature.spectral_centroid(
        y=y,
        sr=sr
    )[0]

    features.update(
        stats(
            centroid,
            "centroid"
        )
    )

    # -----------------------------
    # Bandwidth
    # -----------------------------

    bandwidth = librosa.feature.spectral_bandwidth(
        y=y,
        sr=sr
    )[0]

    features.update(
        stats(
            bandwidth,
            "bandwidth"
        )
    )

    # -----------------------------
    # Rolloff
    # -----------------------------

    rolloff = librosa.feature.spectral_rolloff(
        y=y,
        sr=sr
    )[0]

    features.update(
        stats(
            rolloff,
            "rolloff"
        )
    )

    rms = librosa.feature.rms(
        y=y
    )[0]

    features.update(
        stats(
            rms,
            "rms"
        )
    )

    # -----------------------------
    # ZCR
    # -----------------------------

    zcr = librosa.feature.zero_crossing_rate(
        y
    )[0]

    features.update(
        stats(
            zcr,
            "zcr"
        )
    )

    return features


# =====================================================
# SINGLE WAV
# =====================================================

def extract_wav_features(wav_path):

    try:

        y, sr = librosa.load(
            wav_path,
            sr=16000,
            mono=True
        )

        y, _ = librosa.effects.trim(
            y,
            top_db=30
        )

        peak = np.max(np.abs(y))

        if peak > 0:
            y = y / peak

        features = {}

        features.update(
            extract_librosa_features(
                y,
                sr
            )
        )

        features.update(
            extract_parselmouth_features(
                wav_path
            )
        )

        return features

    except Exception as e:

        print(
            f"Failed: {wav_path}"
        )

        print(e)

        return None


# =====================================================
# SUBJECT LEVEL
# =====================================================

def aggregate_subject(subject_folder):
    all_features = []
    for root, _, files in os.walk(subject_folder):
        for file in files:
            if file.lower().endswith(".wav"):
                path = os.path.join(
                    root,
                    file
                )

                feat = extract_wav_features(
                    path
                )

                if feat is not None:
                    all_features.append(
                        feat
                    )

    if len(all_features) == 0:
        return None

    df = pd.DataFrame(
        all_features
    )

    return df.mean().to_dict()
```

```python
# =====================================================
# MAIN
# =====================================================

ROOT = r"datasets\speech\Italian Parkinson_s Voice and speech"

groups = {
    "15 Young Healthy Control": 0,
    "22 Elderly Healthy Control": 0,
    "28 People with Parkinson's Disease": 1
}

rows = []

for folder_name, label in groups.items():
    group_folder = os.path.join(
        ROOT,
        folder_name
    )

    for participant in tqdm(
        os.listdir(group_folder)
    ):

        participant_path = os.path.join(
            group_folder,
            participant
        )

        print(participant_path)
        if not os.path.isdir(
            participant_path
        ):
            continue

        features = aggregate_subject(
            participant_path
        )

        if features is None:
            continue

        features[
            "participant_name"
        ] = participant

        features[
            "label"
        ] = label

        rows.append(
            features
        )

audio_df = pd.DataFrame(
    rows
)

audio_df.to_csv(
    "audio_features.csv",
    index=False
)

print(audio_df.shape)
print(audio_df.head())
```

```python
"""Italian Parkinson speech pipeline — after master metadata is ready.

What this script does:
1) Loads the master metadata CSV (name, surname, sex, age, time1, CPS1, time2, CPS2, time3, CPS3, label).
2) Walks the audio folders for the 3 groups.
3) Extracts recording-level audio features from every WAV.
4) Aggregates all recordings to one row per participant.
5) Merges the audio features back into the metadata.
6) Saves intermediate and final CSV files.

Expected folder structure (example):
ROOT_AUDIO/
├── 15 Young Healthy Control/
│   ├── Davide S/
│   │   ├── B1....wav
│   │   ├── B2....wav
│   │   └── PR1....wav
│   └── ...
├── 22 Elderly Healthy Control/
│   └── Brigida C/
│       ├── B1....wav
│       ├── B2....wav
│       ├── D1....wav
│       ├── VA1....wav
│       └── ...
└── 28 Parkinson/
    └── ...

Notes:
- The script is robust to extra non-audio files in each participant folder.
- It uses the cleaned audio signal for all Librosa features and the same cleaned signal
  for Parselmouth features, so the feature spaces stay consistent.
- Age is kept as numeric and sex is encoded as M=0, F=1.
"""

from __future__ import annotations

import re
from dataclasses import dataclass
from pathlib import Path
from typing import Dict, Iterable, List, Optional, Tuple

import librosa
import numpy as np
import pandas as pd
import parselmouth
from tqdm import tqdm


# =========================
# CONFIG
# =========================

ROOT_AUDIO = Path(r"datasets\speech\Italian Parkinson_s Voice and speech")
MASTER_METADATA_CSV = Path(r"data.csv")
OUTPUT_DIR = Path(r"italian_pipeline_outputs")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

GROUP_FOLDERS = {
    "15 Young Healthy Control": 0,
    "22 Elderly Healthy Control": 0,
    "28 People with Parkinson's Disease": 1,
}

TARGET_SR = 16000
TRIM_TOP_DB = 25


# =========================
# TEXT / KEY HELPERS
# =========================

def normalize_text(value: object) -> str:
    """Uppercase, remove punctuation/spaces, keep only alnum chars."""
    if pd.isna(value):
        return ""
    text = str(value).strip().upper()
    return re.sub(r"[^A-Z0-9]+", "", text)


def participant_key_from_name_surname(name: object, surname: object) -> str:
    """Create a stable participant key that matches folder naming like 'Davide S'.

    We use the full normalized first name + the first letter of surname.
    Example:
        name='Davide', surname='Sorrenti' -> 'DAVIDES'
        name='Brigida', surname='C'       -> 'BRIGIDAC'
    """
    name_key = normalize_text(name)
    surname_text = normalize_text(surname)
    surname_initial = surname_text[:1] if surname_text else ""
    return f"{name_key}{surname_initial}"


def participant_key_from_folder(folder_name: str) -> str:
    """Normalize a folder name into a comparable participant key."""
    # Example: 'Davide S' -> 'DAVIDES'
    return normalize_text(folder_name)


def infer_task_prefix(filename: str) -> str:
    """Infer the task prefix from a filename.

    Examples:
        B1LBULCAAS94M10...wav -> B1
        PR1BCRAIS...wav       -> PR1
        VA2....wav            -> VA2
    """
    stem = Path(filename).stem.upper()
    m = re.match(r"^([A-Z]{1,3}\d{0,2})", stem)
    if m:
        return m.group(1)
    # Fallback: first 2-3 chars if pattern is unusual
    return stem[:3]


# =========================
# AUDIO PREPROCESSING
# =========================

def load_clean_audio(wav_path: Path) -> Tuple[np.ndarray, int]:
    """Load, clean, trim, and normalize audio."""
    y, sr = librosa.load(str(wav_path), sr=TARGET_SR, mono=True)

    if y.size == 0:
        raise ValueError(f"Empty audio: {wav_path}")

    # Remove DC offset
    y = y - np.mean(y)

    # Trim leading/trailing silence
    y, _ = librosa.effects.trim(y, top_db=TRIM_TOP_DB)

    if y.size == 0:
        raise ValueError(f"Audio became empty after trimming: {wav_path}")

    # Peak normalize
    peak = np.max(np.abs(y))
    if peak > 0:
        y = y / peak

    # Final sanity checks
    if np.isnan(y).any() or np.isinf(y).any():
        raise ValueError(f"NaN/Inf in cleaned signal: {wav_path}")

    if len(y) < int(0.25 * sr):
        raise ValueError(f"Too short after cleaning (<0.25s): {wav_path}")

    return y.astype(np.float32), sr


# =========================
# FEATURE HELPERS
# =========================

def add_stats(features: Dict[str, float], arr: np.ndarray, prefix: str) -> None:
    """Add mean/std/median/min/max stats for an array-like signal."""
    arr = np.asarray(arr)
    arr = arr[np.isfinite(arr)]
    if arr.size == 0:
        features[f"{prefix}_mean"] = np.nan
        features[f"{prefix}_std"] = np.nan
        features[f"{prefix}_median"] = np.nan
        features[f"{prefix}_min"] = np.nan
        features[f"{prefix}_max"] = np.nan
        return

    features[f"{prefix}_mean"] = float(np.mean(arr))
    features[f"{prefix}_std"] = float(np.std(arr))
    features[f"{prefix}_median"] = float(np.median(arr))
    features[f"{prefix}_min"] = float(np.min(arr))
    features[f"{prefix}_max"] = float(np.max(arr))


def safe_parselmouth_features(y: np.ndarray, sr: int) -> Dict[str, float]:
    """Extract pitch/jitter/shimmer/HNR features from the cleaned signal."""
    feats: Dict[str, float] = {}

    try:
        snd = parselmouth.Sound(y, sampling_frequency=sr)

        # Pitch
        pitch = snd.to_pitch()
        pitch_values = pitch.selected_array["frequency"]
        pitch_values = pitch_values[np.isfinite(pitch_values) & (pitch_values > 0)]
        add_stats(feats, pitch_values, "pitch")

        # Point process for perturbation measures
        point_process = parselmouth.praat.call(
            snd,
            "To PointProcess (periodic, cc)",
            75,
            500,
        )

        def praat_scalar(call_args: Tuple, method: str) -> float:
            try:
                return float(parselmouth.praat.call(*call_args, method))
            except Exception:
                return np.nan

        feats["jitter_local"] = praat_scalar(
            (point_process,),
            "Get jitter (local)"
        )
        feats["jitter_rap"] = praat_scalar(
            (point_process,),
            "Get jitter (rap)"
        )
        feats["jitter_ppq5"] = praat_scalar(
            (point_process,),
            "Get jitter (ppq5)"
        )

        # Shimmer requires [snd, point_process]
        try:
            feats["shimmer_local"] = float(
                parselmouth.praat.call(
                    [snd, point_process],
                    "Get shimmer (local)",
                    0,
                    0,
                    0.0001,
                    0.02,
                    1.3,
                    1.6,
                )
            )
        except Exception:
            feats["shimmer_local"] = np.nan

        try:
            feats["shimmer_apq3"] = float(
                parselmouth.praat.call(
                    [snd, point_process],
                    "Get shimmer (apq3)",
                    0,
                    0,
                    0.0001,
                    0.02,
                    1.3,
                    1.6,
                )
            )
        except Exception:
            feats["shimmer_apq3"] = np.nan

        try:
            feats["shimmer_apq5"] = float(
                parselmouth.praat.call(
                    [snd, point_process],
                    "Get shimmer (apq5)",
                    0,
                    0,
                    0.0001,
                    0.02,
                    1.3,
                    1.6,
                )
            )
        except Exception:
            feats["shimmer_apq5"] = np.nan

        # HNR
        try:
            harmonicity = parselmouth.praat.call(
                snd,
                "To Harmonicity (cc)",
                0.01,
                75,
                0.1,
                1.0,
            )
            feats["hnr"] = float(parselmouth.praat.call(harmonicity, "Get mean", 0, 0))
        except Exception:
            feats["hnr"] = np.nan

    except Exception:
        # If Parselmouth fails, return NaNs for the features; do not drop the file
        for col in [
            "pitch_mean", "pitch_std", "pitch_median", "pitch_min", "pitch_max",
            "jitter_local", "jitter_rap", "jitter_ppq5",
            "shimmer_local", "shimmer_apq3", "shimmer_apq5", "hnr",
        ]:
            feats.setdefault(col, np.nan)

    # Make sure all pitch stats exist even if pitch extraction failed
    for col in ["pitch_mean", "pitch_std", "pitch_median", "pitch_min", "pitch_max"]:
        feats.setdefault(col, np.nan)

    return feats


def extract_recording_features(wav_path: Path) -> Optional[Dict[str, object]]:
    """Extract one row of features for one recording."""
    try:
        y, sr = load_clean_audio(wav_path)

        feats: Dict[str, object] = {
            "wav_file": wav_path.name,
            "wav_path": str(wav_path),
            "task": infer_task_prefix(wav_path.name),
            "duration_sec": float(len(y) / sr),
        }

        # MFCCs
        mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
        for i in range(13):
            add_stats(feats, mfcc[i], f"mfcc_{i+1}")

        # Spectral features
        centroid = librosa.feature.spectral_centroid(y=y, sr=sr)[0]
        bandwidth = librosa.feature.spectral_bandwidth(y=y, sr=sr)[0]
        rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)[0]
        rms = librosa.feature.rms(y=y)[0]
        zcr = librosa.feature.zero_crossing_rate(y)[0]

        add_stats(feats, centroid, "centroid")
        add_stats(feats, bandwidth, "bandwidth")
        add_stats(feats, rolloff, "rolloff")
        add_stats(feats, rms, "rms")
        add_stats(feats, zcr, "zcr")

        # Parselmouth / voice quality
        feats.update(safe_parselmouth_features(y, sr))

        return feats

    except Exception as e:
        print(f"[SKIP] {wav_path}: {e}")
        return None


# =========================
# DATASET BUILDING
# =========================

def find_wavs(participant_folder: Path) -> List[Path]:
    """Return all wav files inside a participant folder, recursively."""
    wavs = []
    for p in participant_folder.rglob("*"):
        if p.is_file() and p.suffix.lower() in {".wav", ".wave"}:
            wavs.append(p)
    return sorted(wavs)


def build_recording_table() -> pd.DataFrame:
    """Walk audio folders and create one row per recording."""
    rows: List[Dict[str, object]] = []

    for group_folder_name, label in GROUP_FOLDERS.items():
        group_path = ROOT_AUDIO / group_folder_name
        if not group_path.exists():
            print(f"[WARN] Missing folder: {group_path}")
            continue

        participant_folders = [p for p in group_path.iterdir() if p.is_dir()]
        for participant_folder in tqdm(participant_folders, desc=f"Scanning {group_folder_name}"):
            participant_folder_key = participant_key_from_folder(participant_folder.name)
            wavs = find_wavs(participant_folder)

            for wav_path in wavs:
                feats = extract_recording_features(wav_path)
                if feats is None:
                    continue

                feats["group_folder"] = group_folder_name
                feats["label"] = int(label)
                feats["participant_folder"] = participant_folder.name
                feats["participant_folder_key"] = participant_folder_key
                rows.append(feats)

    recording_df = pd.DataFrame(rows)
    return recording_df


def aggregate_subject_table(recording_df: pd.DataFrame) -> pd.DataFrame:
    """Aggregate all recordings to one row per participant."""
    if recording_df.empty:
        return recording_df.copy()

    # Identify numeric feature columns only; keep metadata columns separately
    meta_cols = {
        "wav_file", "wav_path", "task", "duration_sec", "group_folder", "label",
        "participant_folder", "participant_folder_key",
    }
    numeric_cols = [c for c in recording_df.columns if c not in meta_cols]

    agg_dict = {c: "mean" for c in numeric_cols}
    agg_dict.update({"wav_file": "count"})

    subject_df = (
        recording_df
        .groupby(["participant_folder_key", "participant_folder", "group_folder", "label"], as_index=False)
        .agg(agg_dict)
        .rename(columns={"wav_file": "n_recordings"})
    )

    return subject_df


# =========================
# METADATA MERGE
# =========================

def load_and_prepare_metadata() -> pd.DataFrame:
    """Load master metadata and create a matching participant key."""
    meta = pd.read_csv(MASTER_METADATA_CSV)

    # Normalize expected columns
    rename_map = {}
    for c in meta.columns:
        if c.lower() == "subject_id":
            rename_map[c] = "subject_id"
        elif c.lower() == "cps1":
            rename_map[c] = "CPS1"
        elif c.lower() == "cps2":
            rename_map[c] = "CPS2"
        elif c.lower() == "cps3":
            rename_map[c] = "CPS3"
        elif c.lower() == "time1":
            rename_map[c] = "time1"
        elif c.lower() == "time2":
            rename_map[c] = "time2"
        elif c.lower() == "time3":
            rename_map[c] = "time3"
        elif c.lower() == "sex":
            rename_map[c] = "sex"
        elif c.lower() == "age":
            rename_map[c] = "age"
        elif c.lower() == "label":
            rename_map[c] = "label"
        elif c.lower() == "name":
            rename_map[c] = "name"
        elif c.lower() == "surname":
            rename_map[c] = "surname"

    meta = meta.rename(columns=rename_map)

    required = ["name", "surname", "sex", "age", "time1", "CPS1", "time2", "CPS2", "time3", "CPS3", "label"]
    missing = [c for c in required if c not in meta.columns]
    if missing:
        raise ValueError(f"Missing required metadata columns: {missing}")

    meta["sex"] = meta["sex"].map({"M": 0, "F": 1}).astype("Int64")
    meta["age"] = pd.to_numeric(meta["age"], errors="coerce")
    for c in ["time1", "CPS1", "time2", "CPS2", "time3", "CPS3"]:
        meta[c] = pd.to_numeric(meta[c], errors="coerce")
    meta["label"] = pd.to_numeric(meta["label"], errors="coerce").astype("Int64")

    # Keep raw identity columns but create matching keys
    meta["participant_key"] = meta.apply(
        lambda r: participant_key_from_name_surname(r["name"], r["surname"]),
        axis=1,
    )
    meta["name_norm"] = meta["name"].astype(str).str.strip().str.upper()
    meta["surname_norm"] = meta["surname"].astype(str).str.strip().str.upper()

    return meta


def merge_metadata_and_audio(meta: pd.DataFrame, subject_audio_df: pd.DataFrame) -> Tuple[pd.DataFrame, pd.DataFrame]:
    """Merge metadata with subject-level audio features."""
    if subject_audio_df.empty:
        raise ValueError("No audio features were extracted. Check the audio folder structure and filenames.")

    # Normalize participant folder keys for matching
    audio = subject_audio_df.copy()
    audio["participant_folder_key"] = audio["participant_folder_key"].astype(str).apply(normalize_text)

    # Metadata participant key is name + surname initial, normalized.
    meta = meta.copy()
    meta["participant_key"] = meta["participant_key"].astype(str).apply(normalize_text)

    merged = meta.merge(
        audio,
        left_on="participant_key",
        right_on="participant_folder_key",
        how="left",
        suffixes=("", "_audio"),
    )

    # Rows with no matching audio features
    unmatched = merged[merged["n_recordings"].isna()].copy()

    return merged, unmatched


# =========================
# MAIN
# =========================

def main() -> None:
    print("[1/4] Loading metadata...")
    meta = load_and_prepare_metadata()
    meta.to_csv(OUTPUT_DIR / "metadata_prepared.csv", index=False)

    print("[2/4] Extracting recording-level audio features...")
    recording_df = build_recording_table()
    recording_df.to_csv(OUTPUT_DIR / "recording_features.csv", index=False)

    print("[3/4] Aggregating to subject-level audio features...")
    subject_audio_df = aggregate_subject_table(recording_df)
    subject_audio_df.to_csv(OUTPUT_DIR / "subject_audio_features.csv", index=False)

    print("[4/4] Merging metadata + audio features...")
    final_df, unmatched = merge_metadata_and_audio(meta, subject_audio_df)

    final_path = OUTPUT_DIR / "italian_final_dataset.csv"
    unmatched_path = OUTPUT_DIR / "unmatched_metadata_rows.csv"
    final_df.to_csv(final_path, index=False)
    unmatched.to_csv(unmatched_path, index=False)

    print("\nDone.")
    print(f"Recording rows: {recording_df.shape}")
    print(f"Subject rows:    {subject_audio_df.shape}")
    print(f"Final rows:      {final_df.shape}")
    print(f"Saved: {final_path}")
    print(f"Unmatched rows:  {unmatched_path}")

    # Quick null summary for the final dataset
    nulls = final_df.isnull().sum().sort_values(ascending=False)
    print("\nTop null counts:")
    print(nulls.head(20))


if __name__ == "__main__":
    main()
```
